<h2>Computational scaling of SVMs</h2>

1) Write a python function from which you can generate synthetic datasets of an arbitrary size (e.g. a mixture model of two Gaussians)

2) By training SVMs on datasets of different sizes show:
- how the training time scales.
- how the classification time scales (on a sample of 1000 unseen datapoints).
- the accuracy of the classifier (on a sample of 1000 unseen datapoints).

```python
#import libraries
import pandas as pd
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets.samples_generator import make_blobs

samples = [10**i for i in range(1,6)]

def make_dataset(samples):
    plt.figure(figsize=(20,4))
    for i in range(len(samples)):

        # Generate 2-blob data of size samples[i] each
        X, y = make_blobs(n_samples=samples[i], centers=2, cluster_std=3, random_state = 2)

        plt.subplot(1,5,i+1)
        plt.scatter(X[:, 0], X[:, 1], alpha = 0.5)
        plt.title(f"N = {samples[i]}")
    plt.show()

make_dataset(samples)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

#initiate the linear SVC classifier
svc = SVC(kernel='linear')

def train_test_svm(X_train, X_test, y_train, y_test, demo = False):
    
    #train model
    train_t1_start = time.perf_counter()
    train_t2_start = time.process_time()
    
    model = svc.fit(X_train, y_train) 
    
    train_t1_stop = time.perf_counter()
    train_t2_stop = time.process_time()
    
    #test model
    test_t1_start = time.perf_counter()
    test_t2_start = time.process_time()
    
    score = model.score(X_test, y_test)

    test_t1_stop = time.perf_counter()
    test_t2_stop = time.process_time()
    
    #multiply by 1000 to get milliseconds
    train_t1_time = (train_t1_stop - train_t1_start)*1000
    train_t2_time = (train_t2_stop - train_t2_start)*1000
    test_t1_time = (test_t1_stop - test_t1_start)*1000
    test_t2_time = (test_t2_stop - test_t2_start)*1000
    
    if demo:
        print(f'Train time')
        print(f'Elapsed time: {round(train_t1_time,2)} ms')
        print(f'CPU process time: {round(train_t2_time,2)} ms')

        print(f'\nTest time')
        print(f'Elapsed time: {round(test_t1_time,2)} ms')
        print(f'CPU process time: {round(test_t2_time,2)} ms')

        print(f'Model score: {round(score,4)}')
        
    return train_t1_time, train_t2_time, test_t1_time, test_t2_time, score
    

    
```

```python
#create empty numpy array to store results
results = np.zeros([len(samples),6])

#define a fixed X and y test-set of size=1000 to test the different svc models
X_test, y_test = make_blobs(n_samples=1000, centers=2, cluster_std=3, random_state = 2)

for i in range(len(samples)):
    #create training sets of size samples[i]
    X_train, y_train = make_blobs(n_samples=samples[i], centers=2, cluster_std=3, random_state = 2)
    #get the time taken for training & testing, and the model score
    train_t1_time, train_t2_time, test_t1_time, test_t2_time, score = train_test_svm(X_train, X_test, y_train, y_test)
    #store results
    results[i,:] = samples[i], round(train_t1_time,2), round(train_t2_time,2), round(test_t1_time,2), round(test_t2_time,2), round(score,4)

```

```python
df = pd.DataFrame(results, 
                  columns=['Training size',
                           'Training Elasped Time (ms)', 
                           'Training CPU Processing Time (ms)',
                           'Testing Elasped Time (ms)',
                           'Testing CPU Processing Time (ms)',
                           'Model score (on 1000 samples)'])
df
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.loglog(df['Training size'], df['Training CPU Processing Time (ms)'])
plt.loglog(df['Training size'], df['Testing CPU Processing Time (ms)'])
plt.xlabel('Training Sample Size')
plt.ylabel('CPU Processing Time (ms)')
plt.title('Log plot of CPU processing time (ms) vs SVM Training Sample Size', size=15)
plt.legend()

plt.subplot(1,2,2)
plt.loglog(df['Training size'], df['Model score'])
plt.xlabel('Training Sample Size')
plt.ylabel('Model Score')
plt.title('Log plot of Model Score vs SVM Training Sample Size', size=15)
plt.legend()
plt.tight_layout()
plt.show()

```
